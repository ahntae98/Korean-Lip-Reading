```python
import cv2
import dlib
import numpy as np
import os
a = 1
word = ['마','아','안','애','어','여','오','왠','우','워','으','임']
# 입술을 검출할 동영상 파일 경로
for idx,word in enumerate(word):
    for i in range(40,52,1):
        video_path = '/home/delivery/jongseol/Lipreading/video/video_new/'+word+'_'+str(i)+'.mp4'
        print(video_path)
        output_dir = 'trainset_new/'+word+'/M1/'+str(i)

        # 출력 디렉토리가 없으면 생성
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        # 얼굴 검출기와 얼굴 특징점 검출기 로드
        detector = dlib.get_frontal_face_detector()
        predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

        # 동영상 파일 열기
        cap = cv2.VideoCapture(video_path)
        frame_count = 0
        saved_frame_count = 0

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret or saved_frame_count >= 10:  # 10장을 저장하면 종료
                break

            if frame_count % 3 == 0:
                # 흑백 이미지로 변환
                gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

                # 얼굴 검출
                faces = detector(gray)
                for face in faces:
                    # 얼굴 특징점 검출
                    landmarks = predictor(gray, face)

                    # 입술 좌표 추출
                    lips_points = []
                    for i in range(48, 61):
                        lips_points.append((landmarks.part(i).x, landmarks.part(i).y))

                    # 입술을 포함하는 사각형 영역 계산
                    lips_points = np.array(lips_points)
                    x, y, w, h = cv2.boundingRect(lips_points)

                    # 입술 영역 추출 및 저장
                    lips = frame[y:y+h, x:x+w]
                    output_path = os.path.join(output_dir, f'lips_frame_{saved_frame_count}.jpg')
                    cv2.imwrite(output_path, lips)
                    saved_frame_count += 1

                    if saved_frame_count >= 10:
                        break

            frame_count += 1

        cap.release()
        cv2.destroyAllWindows()
```
